```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
x=np.array(df['year'])
```

```python
x
```

```python
y=np.array(df['income'])
```

```python
def grad(x,y):
    mc=0
    bc=0
    iteration=48
    n=len(x)
    l_rate=0.0001
    for i in range (iteration):
        yp=mc*x+bc
        cost=(1/n)*sum([val**2 for val in (y-yp)])
        md=-(2/n)*sum(x*(y-yp))
        bd=-(2/n)*sum(y-yp)
        mc=mc-l_rate*md
        bc=bc-l_rate*md
        print(f"Iteration={i},m={mc},b={bc},cost={cost}")
    return mc,bc
```

```python
m,b=grad(x,y)
```
