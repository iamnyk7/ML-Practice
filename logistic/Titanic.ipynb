**`so droping the Cabin as it contains many empty values`**

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.head()
```

```python
df_dummies_embarked=pd.get_dummies(df.Embarked)
```

```python
df_dummies_sex=pd.get_dummies(df.Sex)
```

```python
df=pd.concat([df,df_dummies_embarked,df_dummies_sex],axis="columns")
df
```

<h3>Droping columns which are not needed

```python
clean_df=df.drop(['Embarked','Sex','Name','PassengerId','Ticket','Title','Fare'],axis=1)
```

```python
clean_df.head(5)
```

```python
x=clean_df.drop(['Survived'],axis=1)
y=clean_df.Survived
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
