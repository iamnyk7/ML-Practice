```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=load_iris()
df=pd.DataFrame(data.data)
df.columns=data.feature_names
```

```python
df['Species']=data.target
```

```python
df.head()
```

```python
df.shape
```

#  Univariate

```python
df_se=df.loc[df['Species']==0]
df_vg=df.loc[df['Species']==1]
df_vs=df.loc[df['Species']==2]
```

```python
plt.plot(df_se['sepal length (cm)'],np.zeros_like(df_se['sepal length (cm)']),'o')
plt.plot(df_vs['sepal length (cm)'],np.zeros_like(df_vs['sepal length (cm)']),'o')
plt.plot(df_vg['sepal length (cm)'],np.zeros_like(df_vg['sepal length (cm)']),'o')
plt.show()
```

# Bivariate

```python
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,'petal length (cm)','sepal width (cm)').add_legend()
plt.show()
```

# Multivariate

```python
sns.pairplot(df,hue='Species')
```
